```python
import numpy as np
from matplotlib import pyplot as plt

U = 0.01; ##m/s
R = 10000.0; #m
g = 9.81; #m/s2
f = 10**-4; #1/s
r1 = np.arange(0,R+1,1,dtype='int64');
r2 = np.arange(R,3*R,1,dtype='int64');
```

```python
###############Part(a)######################
###cyclonic eddy
Hin_cyca = (U**2+f*U*R)*r1**2/(2*R**2*g);
Hout_cyca = -U**2*R**2/(2*g*r2**2)+U**2/2/g+f*U*R/g*np.log(r2/R)+(U**2+f*U*R)/2/g;
plt.subplots_adjust(wspace=0.6, hspace=0.5)
plt.subplot(1,2,1)
plt.plot(r1,Hin_cyca*100);plt.plot(r2,Hout_cyca*100);
plt.xticks([0,10000,20000,30000],[0,10,20,30])
plt.title('Cyclonic eddy'); 
plt.xlabel('Distance from eddy center [km]'); plt.ylabel(r'$\eta$-$\eta_{center}$[cm]')
###anticyclonic
Hin_anca = (U**2-f*U*R)*r1**2/(2*R**2*g);
Hout_anca = -U**2*R**2/(2*g*r2**2)+U**2/2/g-f*U*R/g*np.log(r2/R)+(U**2-f*U*R)/2/g;
plt.subplot(1,2,2)
plt.plot(r1,Hin_anca*100);plt.plot(r2,Hout_anca*100)
plt.xticks([0,10000,20000,30000],[0,10,20,30])
plt.title('Anticyclonic eddy'); 
plt.xlabel('Distance from eddy center [km]'); plt.ylabel(r'$\eta$-$\eta_{center}$[cm]');
```

```python
###############Part(b)######################
###cyclonic eddy
Hin_cycb = U*f*r1**2/2.0/R/g;
Hout_cycb = U*f*R/2.0/g + f*U*R/g*np.log(r2/R);
###anticyclonic eddy
Hin_ancb = -U*f*r1**2/2.0/R/g;
Hout_ancb = -U*f*R/2.0/g - f*U*R/g*np.log(r2/R);

plt.subplots_adjust(wspace=0.6, hspace=0.5)
plt.subplot(1,2,1)
plt.plot(r1,(Hin_cycb-Hin_cyca)*100);
plt.plot(r2,(Hout_cycb-Hout_cyca)*100);
plt.xticks([0,10000,20000,30000],[0,10,20,30])
plt.title('Cyclonic eddy'); 
plt.xlabel('Distance from eddy center [km]'); plt.ylabel(r'Part(b)-Part(a) [cm]')

plt.subplot(1,2,2)
plt.plot(r1,(Hin_ancb-Hin_anca)*100);
plt.plot(r2,(Hout_ancb-Hout_anca)*100);
plt.xticks([0,10000,20000,30000],[0,10,20,30])
plt.title('Anticyclonic eddy'); 
plt.xlabel('Distance from eddy center [km]'); plt.ylabel(r'Part(b)-Part(a) [cm]');
```
